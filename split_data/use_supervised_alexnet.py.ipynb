```python
import os
num="5"
train_file="/media/liang/ssd2/action_3/model_"+num+"/train"+num+".txt"
test_file="/media/liang/ssd2/action_3/model_"+num+"/test"+num+".txt"
train_file_write="/media/liang/ssd2/action_3/model_"+num+"_supervised/train"+num+".txt"
test_file_write="/media/liang/ssd2/action_3/model_"+num+"_supervised/test"+num+".txt"

read_file=test_file
write_file=test_file_write


f=open(read_file,'r')
f_w= open(write_file,'w')
for line in f.readlines():
    arr=line.strip().split("\t")
    link=arr[0]
    label=arr[1]
    for image in os.listdir("/media/liang/ssd2/action_3/RGB/"+link):
        
        f_w.write("RGB/"+link+"/"+image+" "+label+"\n")#已经有换行了，在label里
        

f.close()
f_w.close()
    
read_file=train_file
write_file=train_file_write


f=open(read_file,'r')
f_w= open(write_file,'w')
for line in f.readlines():
    arr=line.strip().split("\t")
    link=arr[0]
    label=arr[1]
    for image in os.listdir("/media/liang/ssd2/action_3/RGB/"+link):
        f_w.write("RGB/"+link+"/"+image+" "+label+"\n")#已经有换行了，在label里
        

f.close()
f_w.close()
    
```
